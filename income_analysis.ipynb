```python
import pandas as pd
import numpy as np
```

```python
us_zips = pd.read_csv("18zpallagi.csv")
```

```python
us_zips.rename(columns={'A00100':'adj_gross_inc'}, inplace=True)

us_zips
```

```python
us_zips = us_zips[['zipcode','adj_gross_inc']].groupby('zipcode').agg(sum)
us_zips
```

Now that we have all USA zip codes, we need to get a list of only those in NYC. nycbynatives.com offers a complete list of these zip codes along with the corresponding neighborhoods. I copy/pasted them into 'ny_zips.csv'

```python
nyc_zips = pd.read_csv("ny_zips.csv")
# nyc_zips.dropna(axis=1, how='all', inplace=True)

nyc_zips
```

```python
nyc_agi_by_zip = nyc_zips.join(us_zips, how='inner', on='zipcode')

nyc_agi_by_zip
```
